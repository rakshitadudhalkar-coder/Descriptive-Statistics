#9. Combining Measures (Hard)
#A family records their monthly electricity usage (in kWh): [400, 420, 390, 450, 410].Find the mean and standard deviation. What do these values together tell you about the family’s energy use pattern?

Ans -

Data: [400, 420, 390, 450, 410]

Mean = (400+420+390+450+410)/5 = 2070/5 = 414 kWh

Deviations:
(400-414)² = 196

(420-414)² = 36

(390-414)² = 576

(450-414)² = 1296

(410-414)² = 16

Variance = (196+36+576+1296+16)/5 = 2120/5 = 424

SD = √424 ≈ 20.59

These values together tell us that the family's average monthly electricity usage is 414 kWh. The standard deviation of approximately 20.59 kWh indicates that the usage typically deviates by about 20.59 kWh from the mean.
